```python
from sklearn.preprocessing import LabelEncoder

# Initialize LabelEncoder
le = LabelEncoder()

# Select all categorical columns (of type 'object')
cat_columns = df.select_dtypes(include='object').columns

# Apply Label Encoding to each categorical column
df[cat_columns] = df[cat_columns].apply(lambda col: le.fit_transform(col))

# Display the first few rows to verify encoding
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.impute import SimpleImputer

# Separate numeric and categorical columns
num_columns = df.select_dtypes(include=['float64', 'int64']).columns
cat_columns = df.select_dtypes(include=['object']).columns

# Label encode or one-hot encode categorical columns
label_encoder = LabelEncoder()
for col in cat_columns:
    df[col] = label_encoder.fit_transform(df[col])

imputer = SimpleImputer(strategy='mean')
df[num_columns] = imputer.fit_transform(df[num_columns])

print(df.head())
```

```python
# Define the IQR clipping function
def iqr_clipping(df, columns, multiplier=1.5):
    # Copy to avoid modifying original data
    df_clipped = df.copy()

    # Apply IQR clipping for each specified column
    for col in columns:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - multiplier * IQR
        upper_bound = Q3 + multiplier * IQR
        df_clipped[col] = df[col].clip(lower=lower_bound, upper=upper_bound)

    return df_clipped

# List of columns for outlier clipping
outlier_columns = ['Working Professional or Student', 'Work Pressure', 'Job Satisfaction', 'Sleep Duration']

# Apply IQR clipping to the specified columns in df
df_clipped = iqr_clipping(df, outlier_columns, multiplier=1.5)

# Set up plot for visualization
plt.figure(figsize=(15, 8))  # Adjust size for better readability
sns.boxplot(data=df_clipped[outlier_columns])
plt.title("Boxplot After IQR Clipping of Specified Columns")
plt.xticks(rotation=45)
plt.show()
```

```python

from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import LabelEncoder

# Adjust thresholds as per your dataset's context
thresholds = [0.3, 0.7]  # Example thresholds for binning
labels = ["Low", "Medium", "High"]  # Example labels for categories
Y_binned = pd.cut(df_clipped['Depression'], bins=[-float('inf')] + thresholds + [float('inf')], labels=labels)

# Encode categorical target values
label_encoder_y = LabelEncoder()
Y = label_encoder_y.fit_transform(Y_binned)

# Train model with encoded target
X = df_clipped.drop('Depression', axis=1)
model = RandomForestClassifier(random_state=42)
model.fit(X, Y)

# Extract and sort feature importances
feature_importances = model.feature_importances_
feature_importance_df = pd.DataFrame({
    'Feature': X.columns,
    'Importance': feature_importances
}).sort_values(by='Importance', ascending=False)

# Define a threshold for important features and filter
threshold = 0.05
important_features = feature_importance_df[feature_importance_df['Importance'] >= threshold]

# Display important features
print("Important Features (Importance >= 0.05):")
print(important_features)
```

```python
from sklearn.model_selection import train_test_split
from imblearn.ensemble import BalancedBaggingClassifier
from sklearn.tree import DecisionTreeClassifier

# Define features and labels from the DataFrame
Features = df_clipped[important_features['Feature']]
Labels = df_clipped['Depression']

# Convert Labels to integer if it contains continuous values that should represent discrete categories
Labels = Labels.astype(int)

# Split data into training and testing sets
Features_train, Features_test, Labels_train, Labels_test = train_test_split(
    Features, Labels, test_size=0.2, shuffle=True, random_state=42
)

# Initialize the Balanced Bagging Classifier with Decision Tree as the base estimator
balanced_bagging = BalancedBaggingClassifier(
    estimator=DecisionTreeClassifier(),
    sampling_strategy='auto',
    replacement=False,
    random_state=42
)

# Fit the model on the training set
balanced_bagging.fit(Features_train, Labels_train)
```

```python
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, ConfusionMatrixDisplay, roc_auc_score
from sklearn.model_selection import StratifiedKFold, cross_val_score
import matplotlib.pyplot as plt

# Define the XGBClassifier model with parameters
xgb_model = XGBClassifier(
    scale_pos_weight=len(Labels[Labels == 0]) / len(Labels[Labels == 1]),
    reg_lambda=2,
    reg_alpha=1
)

# Train the model
xgb_model.fit(Features_train, Labels_train)

# Make predictions on the test set
train_pred = xgb_model.predict(Features_test)
Pred_proba = xgb_model.predict_proba(Features_test)[:, 1]

# Calculate performance metrics
Training_Accuracy = accuracy_score(Labels_test, train_pred)
Precision = precision_score(Labels_test, train_pred)
Recall = recall_score(Labels_test, train_pred)
F1_score = f1_score(Labels_test, train_pred)
AUC_Score = roc_auc_score(Labels_test, Pred_proba)
cross_val_auc = cross_val_score(xgb_model, Features, Labels, cv=StratifiedKFold(n_splits=5), scoring='roc_auc').mean()

# Display results
print("Training Accuracy:", Training_Accuracy)
print("Precision:", Precision)
print("Recall:", Recall)
print("F1 Score:", F1_score)
print("AUC Score:", AUC_Score)
print("Cross-Validation AUC Score:", cross_val_auc)

# Plot confusion matrix
fig, ax = plt.subplots(figsize=(4, 3))
ConfusionMatrixDisplay(confusion_matrix=confusion_matrix(Labels_test, train_pred),
                       display_labels=['No Depression', 'Depression']).plot(ax=ax, cmap='Blues', values_format='d')
ax.set_title("XGBClassifier")
ax.set_xlabel('Predicted Labels')
ax.set_ylabel('True Labels')
plt.tight_layout()
plt.show()
```
